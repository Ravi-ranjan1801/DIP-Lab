#Lab - 03

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

###Experiment - 01 :Histogram Equalization on a Synthetic Image

```python
#Create a synthetic low-contrast image
synthetic = np.tile(np.arange(256, dtype=np.uint8), (256, 1)) // 2

plt.imshow(synthetic, cmap='gray')
plt.title("Original Synthetic Image")
plt.axis('off')
```

```python
#Apply histogram equalization
equalized = cv2.equalizeHist(synthetic)

plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.imshow(synthetic, cmap='gray')
plt.title("Before Equalization")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(equalized, cmap='gray')
plt.title("After Equalization")
plt.axis('off')
```

###Experiment - 02 : Black & White Images + Arithmetic Operations

```python
#PART A: Horizontal Black & White Lines (Image A)
A = np.zeros((1024, 1024), dtype=np.uint8)

for i in range(0, 1024, 256):
    A[i:i+128, :] = 255   # white lines

plt.imshow(A, cmap='gray')
plt.title("Image A - Horizontal Lines")
plt.axis('off')
```

```python
#PART B: Vertical Black & White Lines (Image B)
B = np.zeros((1024, 1024), dtype=np.uint8)

for j in range(0, 1024, 256):
    B[:, j:j+128] = 255   # white lines

plt.imshow(B, cmap='gray')
plt.title("Image B - Vertical Lines")
plt.axis('off')
```

```python
#Arithmetic Operations on A & B
# (a) Img addition
add = cv2.add(A, B)

plt.imshow(add, cmap='gray')
plt.title("Addition of A and B")
plt.axis('off')
```

```python
#Img Substraction
sub = cv2.subtract(A, B)

plt.imshow(sub, cmap='gray')
plt.title("Subtraction of A and B")
plt.axis('off')
```

```python
#Img Multiplication
mul = cv2.multiply(A, B)

plt.imshow(mul, cmap='gray')
plt.title("Multiplication of A and B")
plt.axis('off')
```

```python
#Img Averaging
avg = cv2.addWeighted(A, 0.5, B, 0.5, 0)

plt.imshow(avg, cmap='gray')
plt.title("Average of A and B")
plt.axis('off')
```

###Experiment - 03 : Sinusoidal Intensity Image

```python
#Generate sinusoidal image
rows, cols = 256, 1024
x = np.linspace(0, 2*np.pi, cols)
sin_wave = (127.5 * (1 + np.sin(x))).astype(np.uint8)

sin_image = np.tile(sin_wave, (rows, 1))

plt.imshow(sin_image, cmap='gray')
plt.title("Sinusoidal Grayscale Image")
plt.axis('off')
```

###Experiment - 04 : White Image with Black Box at Center

```python
img = np.ones((256, 256), dtype=np.uint8) * 255

start = (256 - 58) // 2
end = start + 58

img[start:end, start:end] = 0

plt.imshow(img, cmap='gray')
plt.title("White Image with Black Center Box")
plt.axis('off')
```
