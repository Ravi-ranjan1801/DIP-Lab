```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("image.jpg")
```

```python
# Read image
img = cv2.imread("image.jpg")   # put your image name here

# Convert BGR to RGB for correct display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display image
plt.imshow(img_rgb)
plt.title("Original Color Image")
plt.axis("off")

# Write image in different formats
cv2.imwrite("output.png", img)
cv2.imwrite("output.tiff", img)
```

```python
# Split channels
B, G, R = cv2.split(img)

plt.figure(figsize=(10,4))

plt.subplot(1,3,1)
plt.imshow(R, cmap='gray')
plt.title("Red Plane")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(G, cmap='gray')
plt.title("Green Plane")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(B, cmap='gray')
plt.title("Blue Plane")
plt.axis("off")
```

```python
merged = cv2.merge([B, G, R])
merged_rgb = cv2.cvtColor(merged, cv2.COLOR_BGR2RGB)

plt.imshow(merged_rgb)
plt.title("Reconstructed Color Image")
plt.axis("off")
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")
```

```python
# Thresholding
_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

plt.imshow(binary, cmap='gray')
plt.title("Binary Image")
plt.axis("off")
```

```python
half = cv2.resize(img, None, fx=0.5, fy=0.5)
quarter = cv2.resize(img, None, fx=0.25, fy=0.25)

plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(half, cv2.COLOR_BGR2RGB))
plt.title("Half Size")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(quarter, cv2.COLOR_BGR2RGB))
plt.title("Quarter Size")
plt.axis("off")
```

```python
(h, w) = img.shape[:2]
center = (w//2, h//2)

# 45 degree
M45 = cv2.getRotationMatrix2D(center, 45, 1.0)
rot45 = cv2.warpAffine(img, M45, (w, h))

# 90 degree
rot90 = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)

# 180 degree
rot180 = cv2.rotate(img, cv2.ROTATE_180)

plt.figure(figsize=(10,4))

plt.subplot(1,3,1)
plt.imshow(cv2.cvtColor(rot45, cv2.COLOR_BGR2RGB))
plt.title("45 Degree")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(cv2.cvtColor(rot90, cv2.COLOR_BGR2RGB))
plt.title("90 Degree")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(cv2.cvtColor(rot180, cv2.COLOR_BGR2RGB))
plt.title("180 Degree")
plt.axis("off")
```
