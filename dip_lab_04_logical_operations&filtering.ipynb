#**Lab - 04**



```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

##Experiment - 01 : LOGICAL OPERATIONS

```python
#create 2 grayscale images
# Create two 256x256 grayscale images
img1 = np.zeros((256,256), dtype=np.uint8)
img2 = np.zeros((256,256), dtype=np.uint8)

# Draw shapes
cv2.rectangle(img1, (50,50), (200,200), 255, -1)
cv2.circle(img2, (128,128), 80, 255, -1)

plt.figure(figsize=(6,3))
plt.subplot(1,2,1)
plt.imshow(img1, cmap='gray')
plt.title("Image 1")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(img2, cmap='gray')
plt.title("Image 2")
plt.axis('off')
```

###a) AND Operation

```python
and_img = cv2.bitwise_and(img1, img2)

plt.imshow(and_img, cmap='gray')
plt.title("AND Operation")
plt.axis('off')
```

###b) OR Operation

```python
or_img = cv2.bitwise_or(img1, img2)

plt.imshow(or_img, cmap='gray')
plt.title("OR Operation")
plt.axis('off')
```

###c) XOR Operation

```python
xor_img = cv2.bitwise_xor(img1, img2)

plt.imshow(xor_img, cmap='gray')
plt.title("XOR Operation")
plt.axis('off')
```

###d) NOT Operation

```python
not_img = cv2.bitwise_not(img1)

plt.imshow(not_img, cmap='gray')
plt.title("NOT Operation")
plt.axis('off')
```

#Experiment - 02  SALT & PEPPER NOISE + FILTERING

```python
# take grayscale image
img = cv2.imread('gray.jpeg', cv2.IMREAD_GRAYSCALE)

plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis('off')
```

```python
# add salt & pepper noise
noisy = img.copy()
prob = 0.02

# Salt
salt = np.random.rand(*img.shape) < prob
noisy[salt] = 255

# Pepper
pepper = np.random.rand(*img.shape) < prob
noisy[pepper] = 0

plt.imshow(noisy, cmap='gray')
plt.title("Salt & Pepper Noise")
plt.axis('off')
```

###a) Mean Filtering

```python
mean_filter = cv2.blur(noisy, (3,3))

plt.imshow(mean_filter, cmap='gray')
plt.title("Mean Filter")
plt.axis('off')
```

###b) Weighted Mean Filtering (Gaussian)

```python
gaussian = cv2.GaussianBlur(noisy, (5,5), 0)

plt.imshow(gaussian, cmap='gray')
plt.title("Weighted Mean (Gaussian)")
plt.axis('off')
```

###c) Median Filtering

```python
median = cv2.medianBlur(noisy, 3)

plt.imshow(median, cmap='gray')
plt.title("Median Filter")
plt.axis('off')
```

#PART 3  SHARPENING OPERATIONS

###a) Laplacian Filter

```python
laplacian = cv2.Laplacian(img, cv2.CV_64F)
laplacian = np.uint8(np.absolute(laplacian))

plt.imshow(laplacian, cmap='gray')
plt.title("Laplacian")
plt.axis('off')
```

###b) Unsharp Masking

```python
blur = cv2.GaussianBlur(img, (5,5), 0)
mask = cv2.subtract(img, blur)
unsharp = cv2.add(img, mask)

plt.imshow(unsharp, cmap='gray')
plt.title("Unsharp Masking")
plt.axis('off')
```

###c) High Boost Filtering

```python
k = 2
high_boost = cv2.add(img, k*mask)

plt.imshow(high_boost, cmap='gray')
plt.title("High Boost Filtering")
plt.axis('off')
```
